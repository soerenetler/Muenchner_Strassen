```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("raw_data/vablock_stadtbezirk.geojson")
gdf = gdf.set_index("sb_nummer")
gdf = gdf.set_crs(epsg=25832)
gdf = gdf.drop(columns=["id", "flaeche_qm"])
gdf = gdf.dissolve(by="sb_nummer", as_index=False)
gdf
```

```python
gdf.plot()
```

```python
gdf.to_file("processed_data/muenchen_stadtbezirke.geojson", driver="GeoJSON")
```

```python
gdf_wgs84 = gdf.to_crs(epsg=4326)
gdf_wgs84.to_file("processed_data/muenchen_stadtbezirke_wgs84.geojson", driver="GeoJSON")
```
